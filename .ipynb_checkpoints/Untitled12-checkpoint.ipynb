```python
out = []  # Global list

def out_det(jj):
    if not pd.api.types.is_numeric_dtype(jj):
        return  # Skip non-numeric columns

    q1 = jj.quantile(0.25)
    q3 = jj.quantile(0.75)
    iqr = q3 - q1

    # Find outliers using vectorized filtering
    outliers = jj[(jj > (q3+1.5*iqr)) | (jj < (q1 - 1.5*iqr))]
    
    out.extend(outliers.tolist())  # ✅ Append only the outlier values

# Apply function to all columns
for col in heart.columns:
    out_det(heart[col])

print(out)  # ✅ List of all outliers
```

```python
out = []  # Global list

def out_det(jj):
    if not pd.api.types.is_numeric_dtype(jj):
        return
    
    q1 = jj.quantile(0.25)
    q3 = jj.quantile(0.75)
    iqr = q3 - q1

    for i in jj:
        if i > q3 + 1.5 * iqr or i < q1 - 1.5 * iqr:
            out.append(i)  # Appending to the global list
            #A Pandas Series is never directly True or False in an if condition. So, we did not use heart[] in if 

for col in heart.columns:
    out_det(heart[col])

print(out)  # ❌ Contains outliers from ALL columns combined
```

```python
# Since having a global list is not a good idea, we write it in another way:
def out_det(jj):
    if not pd.api.types.is_numeric_dtype(jj):
        return []
    
    q1 = jj.quantile(0.25)
    q3 = jj.quantile(0.75)
    iqr = q3 - q1

    # Find outliers
    outliers = jj[(jj > q3 + 1.5 * iqr) | (jj < q1 - 1.5 * iqr)]
    
    return outliers.tolist()  # ✅ Returns a fresh list each time

# Store outliers in a dictionary per column
outlier_dict = {}

for col in heart.columns:
    outliers = out_det(heart[col])
    if outliers:  # Only store if there are outliers
        outlier_dict[col] = outliers

print(outlier_dict)
```

```python
out1 = []
def out_zscore(jj):
    if jj.dtype not in ("int64", "float64"):
        return

    l = jj.mean() - 3*jj.std()
    u = jj.mean() + 3*jj.std()
    
    for i in jj:
        if (i > u ) | (i < l):
            out1.append(i)
    
for col in heart.columns:
    outliers = out_zscore(heart[col])

print(out1)
```

```python
def out_z(jj):
    if not pd.api.types.is_numeric_dtype(jj):
        return []
    
    l = jj.mean() - 3*jj.std()
    u = jj.mean() + 3*jj.std()

    # Find outliers
    outliers = jj[(jj > u) | (jj < l)]
    
    return outliers.tolist()  # ✅ Returns a fresh list each time

# Store outliers in a dictionary per column
outlier_dict = {}

for col in heart.columns:
    outliers = out_z(heart[col])
    if outliers:  # Only store if there are outliers
        outlier_dict[col] = outliers

print(outlier_dict)
```
