```python


def outliers_def(column_name, column_label):

    Q1 = column_name.quantile(.25)
    Q3 = column_name.quantile(.75)
    IQR = Q3 - Q1
    L = Q1 - 1.5*IQR
    U = Q3 + 1.5*IQR

    
    out = []
#    out = [i for i in column_name if i < L or i > U]  # List comprehension (faster)
    
    for i in column_name:
        if i < L or i > U:
            out.append(i)
    print(f'For column "{column_label}", outliers are: {out}')  

for j in df.select_dtypes(include=['number']).columns:
    outliers_def(df[j], j)
    
```

```python

def zscore_outliers(col):
    z = ((col - col.mean()) / col.std()).abs()
    return z > 3  # Returns a boolean Series

# Create a boolean DataFrame of the same shape as df[numeric columns]
outlier_flags = pd.DataFrame(False, index=df.index, columns=df.select_dtypes(include=['number']).columns)

for column in df.select_dtypes(include=['number']).columns:
    outlier_flags[column] = zscore_outliers(df[column])

# Get rows where ANY column is an outlier
rows_with_outliers = outlier_flags.any(axis=1)

# Separate the rows
df_outliers = df[rows_with_outliers]        # Rows with outliers
df_cleaned = df[~rows_with_outliers]  
```

```python
import numpy as np
import pandas as pd
from scipy import stats

def outlier(column_name, col_label):
    """Detects outliers using IQR and Z-score methods."""
    
    column_name = column_name.dropna()
    
    # IQR Method
    q1 = column_name.quantile(0.25)
    q3 = column_name.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers_iqr = column_name[(column_name < lower_bound) | (column_name > upper_bound)].tolist()

    # Z-score Method
    if column_name.std() == 0:
        z_scores = np.zeros_like(column_name)
    else:
        z_scores = np.abs(stats.zscore(column_name))
        
    outliers_z = column_name.loc[z_scores > 3].tolist()

    print(f"Column '{col_label}' - min: {column_name.min()}, max: {column_name.max()}, mean: {column_name.mean()}, median: {column_name.median()}")
    print(f"For IQR method, outliers are: {outliers_iqr}")
    print(f"For Z-score method, outliers are: {outliers_z}\n")

# Clean column names
df.columns = df.columns.str.strip()

# Apply function to numeric columns, ignoring categorical & binary variables
for col in df.select_dtypes(include=[np.number]).columns:
    if df[col].nunique() > 2:  # Exclude binary & categorical-like numeric columns
        outlier(df[col], col)
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X = df_n.drop('c', axis=1)
y = df_n['c']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =0.25, random_state = 7)

k_range = list(range(1, 26))
scores = []

for i in k_range:
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(y_test, y_pred))

plt.plot(k_range, scores)
plt.xlabel('Values of k for KNN')
plt.ylabel('Accuracy')
plt.title('Accuracy Scores for values of K  of K-Nearst-Neighbors')
plt.show()

```

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
scores = cross_val_score(knn, X, y, cv = 5, scoring = 'accuracy')
scores
```
