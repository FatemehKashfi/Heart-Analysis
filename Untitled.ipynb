```python
def outliers(column):
    
    if column.dtype not in ['int', 'float']:
        return
        
    q1 = column.quantile(.25)
    q3 = column.quantile(.75)
    outs = column[(column < (q1-1.5*(q3-q1)))|(column>(q3+1.5*(q3-q1)))]
    print(f'column: {column.name}, {list(outs)}') 

for col in heart.columns:
    outliers(heart[col])
```
