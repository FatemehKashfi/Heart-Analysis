```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/mahsa/Downloads/ipynbs/Heart data.csv")
df
```

```python
df = df.fillna(0)
```

```python
x = df.drop('c', axis =1)
y = df.c
```

```python
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(
    x, y, test_size = 0.25, random_state = 0
)
k_range = list(range(1, 26))
scores = []
for i in k_range:
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    scores.append(metrics.accuracy_score(Y_test, Y_pred))
```

```python
plt.plot(k_range, scores)
plt.xlabel('value of k for knn')
plt.ylabel('Accuracy score')
plt.title('Accurecy scores for values of k of k-nearest Neighbors')
plt.show()
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
```

```python
KNN = KNeighborsClassifier()
scores = cross_val_score(KNN, x, y, scoring = 'accuracy')
scores
```

```python
k_range = range(1, 10)
weights_custom =['uniform', 'distance']
param_grid = dict(n_neighbors = k_range, weights = weights_custom)
param_grid
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
KNN_Grid = GridSearchCV(KNN, param_grid, cv = 5, scoring = 'accuracy')
```

```python
KNN_Grid.fit(x,y)
```
